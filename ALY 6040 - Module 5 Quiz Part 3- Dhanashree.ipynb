```python
#GOT_new=GOT[['Gender','Nobility','FfC','DwD','Martell','Tyrell','Status']]
#X_new= LifeExpectancy[['adult_mortality','infant_deaths']]
GOT_new=GOT[['Gender','Nobility','FfC','DwD','Martell','Tyrell','Status']]
```

## From the corelation matrix, got the columns which are significantly helping to predict the target variable

```python
GOT_new
```

```python
X=GOT_new.iloc[:,:-1].values
y =GOT_new.iloc[:, -1].values
```

## Test-Train Split 40%

```python
#Creating train-test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.40, random_state = 0)
```

## Building logistic regression model

```python
## Training the Logistic Regression model on the Training set
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
#Checking accuracy of model
from sklearn.metrics import confusion_matrix, accuracy_score
accuracy_score(y_test, y_pred)
```

```python
#Cross-verify labels
pd.crosstab(y_test, y_pred, rownames=['True'], colnames=['Predicted'], margins=True)
```

## Got the accuracy of 66% and from the confusion matrix we can analyze that that 55 values have been correctly labelled as status “Dead” whereas 188 values have been correctly labelled as status “Alive”, but 50 values which were actually classified as “Alive” were incorrectly classified by model as “Dead” Also, 74 values which were actually classified as “Dead” have been incorrectly labelled as “Alive”

## Decision Tree

```python
# decision tree model
from sklearn.tree import DecisionTreeClassifier
dtree_model=DecisionTreeClassifier(random_state=200)
```

```python
# Create a dictionary of all values we want to hypertune
param_grid = {'criterion':['gini','entropy'],'max_depth': np.arange(3, 15),'min_samples_split': [ 2, 3, 4]}
```

```python
#Use gridsearch to test all values
from sklearn.model_selection import GridSearchCV

grid_search_cv = GridSearchCV(dtree_model, param_grid, cv=10)
grid_search_cv.fit(X_train, y_train)
```

```python
#Get the best parameters
grid_search_cv.best_params_
```

```python
#Creating decision tree model with best parameters
dt=DecisionTreeClassifier(criterion='gini',max_depth=4,min_samples_split=2)
dt.fit(X_train, y_train)
y_pred = dt.predict(X_test)
```

```python
# Creating Decision Tree tree model with best parameters
import sklearn.metrics as metrics

acc_rf = metrics.accuracy_score(y_pred, y_test)
print('The accuracy of the Decision Tree is', acc_rf)
```

```python
#Cross-verify labels
pd.crosstab(y_test, y_pred, rownames=['True'], colnames=['Predicted'], margins=True)
```

## Got the accuracy of 67% and from the confusion matrix we can analyze that that 84 values have been correctly labelled as status “Dead” whereas 159 values have been correctly labelled as status “Alive”, but 79 values which were actually classified as “Alive” were incorrectly classified by model as “Dead” Also, 45 values which were actually classified as “Dead” have been incorrectly labelled as “Alive”

## Comparion of models using ROC and AUC Curve, hence we can conclude that decision tree model worked better for this dataset as compared to logistic regression

```python
from sklearn import metrics
import numpy as np
import matplotlib.pyplot as plt

plt.figure(0).clf()

pred = classifier.predict(X_test)
label = y_test
fpr, tpr, thresh = metrics.roc_curve(label, pred)
auc = metrics.roc_auc_score(label, pred)
plt.plot(fpr,tpr,label="Logistic Regression, auc="+str(auc))


pred = dt.predict(X_test)
label = y_test
fpr, tpr, thresh = metrics.roc_curve(label, pred)
auc = metrics.roc_auc_score(label, pred)
plt.plot(fpr,tpr,label="Decision Tree, auc="+str(auc))



plt.legend(loc=0)
```
